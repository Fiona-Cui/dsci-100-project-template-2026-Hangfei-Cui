```r
library(tidyverse)
library(tidymodels)
set.seed(123)
# convert tournament results into factors (categorical variables)
tidy_data_2 <- tidy_data|>
mutate(tourney_finish = as.factor(tourney_finish))
# Split the tidy_data into training set(75%) and testing set(25%). Stratified by tournament_finish.
data_split <- initial_split(tidy_data_2, prop = 0.75, strata = tourney_finish)
data_train <- training(data_split)
data_test <- testing(data_split)
# Create a recipe, center and scale predictors for KNN on training data.
knn_recipe <- recipe(tourney_finish ~ seed + reg_wins_pct, data = data_train)|>
step_center(all_predictors())|>
step_scale(all_predictors())
# Create KNN classification model
knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune())|>
set_engine("kknn")|>
set_mode("classification")
# Create 5-fold cross-validation on training data
data_vfold <- vfold_cv(data_train, v = 5, strata = tourney_finish)
# Create grid of k values for tuning.
k_vals <- tibble(neighbors = seq(from = 1, to = 200, by = 3))
# Fit the model and test the performances of different K
data_fit_2 <- workflow()|>
add_recipe(knn_recipe)|>
add_model(knn_spec)|>
tune_grid(resamples = data_vfold, grid = k_vals)
# Collect metrics and filter the data_fit_2 to only keep accuracy metrics.               
accuracies <- data_fit_2|>
collect_metrics()|>
filter(.metric == "accuracy")
head(accuracies)
```
